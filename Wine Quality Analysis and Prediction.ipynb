# Data Filling for NaN values

```python
df.isnull().sum()
```

```python
df['fixed acidity'] = df['fixed acidity'].fillna(df['fixed acidity'].mean())
df['volatile acidity'] = df['volatile acidity'].fillna(df['volatile acidity'].mean())
df['citric acid'] = df['citric acid'].fillna(df['citric acid'].mean())
df['residual sugar'] = df['residual sugar'].fillna(df['residual sugar'].mean())
df['chlorides'] = df['chlorides'].fillna(df['chlorides'].mean())
df['pH'] = df['pH'].fillna(df['pH'].mean())
df['sulphates'] = df['sulphates'].fillna(df['sulphates'].mean())
```

```python
df.isnull().sum()
```

All null values are replaced with their respective mean because all columns which were containing null values are of float64 type

```python
df_copy = df.copy()
df_copy.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df_copy['type'] = lb.fit_transform(df_copy['type'])
```

```python
df_copy.head()
```

```python
df_copy["quality"] = df_copy["quality"].astype(int)
```

```python
df_copy.head()
```

```python
quaity_mapping = { 3 : "Low",4 : "Low",5: "Medium",6 : "Medium",7: "Medium",8 : "High",9 : "High"}
df_copy["quality"] =  df_copy["quality"].map(quaity_mapping)
```

```python
lower_limit = df_copy["free sulfur dioxide"].mean() - 3*df_copy["free sulfur dioxide"].std()
upper_limit = df_copy["free sulfur dioxide"].mean() + 3*df_copy["free sulfur dioxide"].std()

df_copy = df_copy[(df_copy["free sulfur dioxide"] > lower_limit) & (df_copy["free sulfur dioxide"] < upper_limit)]

lower_limit = df_copy['total sulfur dioxide'].mean() - 3*df_copy['total sulfur dioxide'].std()
upper_limit = df_copy['total sulfur dioxide'].mean() + 3*df_copy['total sulfur dioxide'].std()
print(lower_limit,upper_limit)

df_copy = df_copy[(df_copy['total sulfur dioxide'] > lower_limit) & (df_copy['total sulfur dioxide'] < upper_limit)]
df_copy.head()

lower_limit = df_copy['residual sugar'].mean() - 3*df_copy['residual sugar'].std()
upper_limit = df_copy['residual sugar'].mean() + 3*df_copy['residual sugar'].std()
print(lower_limit,upper_limit)

df_copy = df_copy[(df_copy['residual sugar'] > lower_limit) & (df_copy['residual sugar'] < upper_limit)]
df_copy.head()
```

```python
mapping_quality = {"Low" : 0,"Medium": 1,"High" : 2}
df_copy["quality"] =  df_copy["quality"].map(mapping_quality)
```

# Model Building

## Random Forest

```python
X = df_copy.drop(['quality'], axis=1)
y = df_copy['quality']
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=32)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

rf = RandomForestClassifier(n_estimators=100, max_depth=20, random_state=40)
rf.fit(X_train, y_train)
```

```python
score = rf.score(X_train, y_train)
score2 = rf.score(X_test, y_test)
score, score2
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier(criterion='entropy', min_samples_split=200)

dt_clf.fit(X_train, y_train)

socre = accuracy_score(y_train, dt_clf.predict(X_train)) #training accuracy
score1 = accuracy_score(y_test, dt_clf.predict(X_test)) #testing accuracy

score, score1
```

## KNN Classifier

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler().fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=17)
knn.fit(X_train_std,y_train)
```

```python
#prediction of the KNN model
predict_knn = knn.predict(X_test_std)
accuracy_knn = accuracy_score(y_test,predict_knn)*100
print("Accuracy of the KNN classifier is:",accuracy_knn) #testing accuracy
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression 
log_clf = LogisticRegression(random_state = 0) 
log_clf.fit(X_train_std,y_train) 
```

```python
socre = accuracy_score(y_train, log_clf.predict(X_train_std)) #training accuracy
score1 = accuracy_score(y_test, log_clf.predict(X_test_std)) #testing accuracy

score, score1
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
model_params  = {
    "svm" : {
        "model":SVC(gamma="auto"),
        "params":{
            'C' : [1,10,20],
            'kernel':["rbf"]
        }
    },
    
    "decision_tree":{
        "model": DecisionTreeClassifier(),
        "params":{
            'criterion':["entropy","gini"],
            "max_depth":[5,8,9]
        }
    },
     "random_forest":{
        "model": RandomForestClassifier(),
        "params":{
            "n_estimators":[1,5,10],
            "max_depth":[5,8,9]
        }
    },
    "naive_bayes":{
        "model": GaussianNB(),
        "params":{}
    },
    
    'logistic_regression' : {
        'model' : LogisticRegression(solver='liblinear',multi_class = 'auto'),
        'params': {
            "C" : [1,5,10]
        }
    }
    
}
```

```python
score=[]
for model_name,mp in model_params.items():
    clf = GridSearchCV(mp["model"],mp["params"],cv=8,return_train_score=False)
    clf.fit(X,y)
    score.append({
        "Model" : model_name,
        "Best_Score": clf.best_score_,
        "Best_Params": clf.best_params_
    })
df_scores = pd.DataFrame(score,columns=["Model","Best_Score","Best_Params"])
```
